```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Step 2: Load dataset
df = pd.read_csv('AirPassengers.csv')
print("First 5 rows:\n", df.head())
print("\nShape of dataset:", df.shape)

# Step 3: Convert 'Month' to datetime and set as index
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Step 4: Visualize the data
plt.figure(figsize=(10, 4))
plt.plot(df, label='Monthly Air Passengers')
plt.title('Monthly Air Passengers from 1949 to 1960')
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
# Changed 'Passengers' to '#Passengers' to match column name in DataFrame
scaled_data = scaler.fit_transform(df[['#Passengers']])

# Step 6: Create sequences
def create_sequences(data, time_steps=12):
    X, y = [], []
    for i in range(time_steps, len(data)):
        X.append(data[i - time_steps:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

time_steps = 12
X, y = create_sequences(scaled_data, time_steps)

# Reshape X to fit LSTM input (samples, time_steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 7: Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Step 8: Build LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 9: Train the model
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=0)

# Step 10: Make predictions
predictions = model.predict(X_test)
predicted_passengers = scaler.inverse_transform(predictions.reshape(-1, 1))
actual_passengers = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 11: Plot actual vs predicted
plt.figure(figsize=(10, 4))
plt.plot(actual_passengers, label='Actual')
plt.plot(predicted_passengers, label='Predicted')
plt.title('Air Passenger Forecasting with LSTM')
plt.xlabel('Time Step')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()
```
